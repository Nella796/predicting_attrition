```python
df.head()
```

```python
# I wrote this function to create stacked barchart visualizations for each categorical variable.
def stacked_bar(df, cat1, cat2, condition, width = .35, rotation = 45, label1 = '', label2 = ''):
    df1 = df[df[cat2] == condition]
    df2 = df[df[cat2] != condition]
    
    dict1 = df1[cat1].value_counts().to_dict()
    dict2 = df2[cat1].value_counts().to_dict()
    
    key1 = dict1.keys()
    key2 = dict2.keys()
    
    value1 = list(dict1.values())
    value2 = list(dict2.values())
    
    figure = plt.figure()
        
    plt.bar(key1, value1, width = width, color = 'blue', label = 'Non Attrition')
    plt.bar(key2, value2, width = width, bottom = value1, color = 'red', label = 'Attrition')
    plt.xlabel = 'Department'
    plt.ylabel = 'Respondents'
    plt.xticks(rotation = rotation)
    plt.legend()
    
    
    return(figure)
```

```python
# After running into issues with the ordering in my previous function I created this modified version to fix previous issues.
def stacked_bar_modified(df, cat1, cat2, condition, width = .35, rotation = 45, label1 = '', label2 = ''):
    df1 = df[df[cat2] == condition]
    df2 = df[df[cat2] != condition]
    
    dict1 = df1[cat1].value_counts().to_dict()
    dict2 = df2[cat1].value_counts().to_dict()
    
    key1 = sorted(dict1)
    key2 = sorted(dict2)
    
    value1 = []
    value2 = []
    
    for value in key1:
        value1.append(dict1[value])
    for value in key2:
        value2.append(dict2[value])
    
    figure = plt.figure()
    
    plt.bar(key1, value1, width = width, color = 'blue', label = 'Non Attrition')
    plt.bar(key2, value2, width = width, bottom = value1, color = 'red', label = 'Attrition')
    plt.xlabel = 'Department'
    plt.ylabel = 'Respondents'
    plt.xticks(rotation = rotation)
    plt.legend()
    
    
    return(figure)
```

# Visualizing P Values

```python
cross_tab_list = [performance_cross, satisfaction_cross, role_cross, involvement_cross, environment_cross, Worklife_cross, department_cross, education_cross, field_cross, relationship_cross]
```

```python
p_values = dict()
for cross_tab in cross_tab_list:
    p_values[cross_tab.index.name] = chi2_contingency(cross_tab)[1]
```

```python
p_values
```

##### Lower P values indicate larger statistical significance

```python
feature = p_values.keys()
values = p_values.values()
```

```python
plt.bar(feature,values)
plt.xticks(rotation='vertical')
plt.title('p_values (unscaled)')
```

```python
plt.bar(feature,values)
plt.ylim([0,.008])
plt.xticks(rotation='vertical')
plt.title('p_values (scaled to .008)')
```

```python
plt.bar(feature,values, color=['C0', 'red', 'C0', 'C0', 'C0', 'C0', 'C0','C0','C0','C0'])
plt.ylim([0,.05])
plt.xticks(rotation='vertical')
plt.title('p_values (scaled to .008)')
plt.savefig('Images\P_values_scaled_008_Job_satisfaction', bbox_inches = 'tight')
```

# Conclusion

The p values of the Chi squared tests were significant at the 95% level for all the tested variables except for the relationship satisfaction, performance rating, and the education level variables. This indicates that the ratios of the remaining variables are abnormal, and it is possible that they have the explanatory power to predict attrition. Job Role, Job Involvement, and Environment Satisfaction were highly significant and were the only variables that had p values less than .0005. While statistical significance might be a sign that the features will be useful in prediction models, it isn't guaranteed. There is also a possibility that these features pick up on similar indicators. For instance, if job satisfaction happens to be reliant on environment satisfaction in a significant capacity, including both features in the model would be redundant and inefficient. Ultimately deciding which features are best suited for prediction will be decided in the feature engineering stage.
